```python
# importing the libraries
import pandas as pd
import statsmodels.api as sm
```

```python
# importing the dataset
dataset = pd.read_csv('pp-complete.csv')
```

```python
# A random 50% sample of the data with replacement
df = dataset.sample(frac=0.5, replace=True, random_state=1)
```

```python
# Selecting features
df = df.rename(columns={'18500':'price', '1995-01-31 00:00': 'date', 'F': 'property', 'L':'duration', 'TORQUAY.1': 'town'})
df = df[['price', 'date', 'property', 'duration', 'town']]
df = df[df['town'].str.contains("LONDON")]
df['date'] = pd.to_datetime(df['date'])  
```

```python
# Encoding categorical data
X = pd.get_dummies(df.iloc[:,2:],drop_first=True)
Y = df.iloc[:,0]
```

```python
# Splitting the dataset into the Training set and Test set
train_mask = (df['date'] <= '2016-1-1')
test_mask = (df['date'] > '2016-1-1')

X_train = X.loc[train_mask]
X_test = X.loc[test_mask]

y_train = Y.loc[train_mask].values
y_test = Y.loc[test_mask].values
```

```python
# Fitting Multiple Linear Regression to the Training set
results = sm.OLS(y_train, X_train).fit()
print(results.summary())
```

## Summary of the model and evaluation result

To predict dependent variable (sale prices) which will vary based on the value of the independent variable. Multiple linear regression was used to model. The results are shown in the figure. R-squared value r < 0.5 which is moderate effect size. The predictor variables of property are significant because both of their p-values are 0.000. However, the p-value for duration_u is greater than the common alpha level of 0.05, which indicates that it is not statistically significant.
